## Introduction

In this notebook we will plot the time-step evolutions during the performed simulations to see whether they reflect the system's physical dynamics or not. Specifically, we will compare the time-step evolutions during simulations performed with different grid resolutions and different time-stepping strategies (with and without the normalization factor).

```python
# Import the necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Import the csv file with the data from the simulation
data_3ref = pd.read_csv('data/unnorm_level2/time_steps.csv')
data_2ref = pd.read_csv('data/unnorm_level1/time_steps.csv')
data_norm = pd.read_csv('data/norm_level2/time_steps.csv')

# Remove the first row of the data frames
data_3ref = data_3ref.iloc[1:]
data_2ref = data_2ref.iloc[1:]
data_norm = data_norm.iloc[1:]

# Add labels to the columns
data_3ref.columns = ['step', 'time-step', 'time']
data_2ref.columns = ['step', 'time-step', 'time']
data_norm.columns = ['step', 'time-step', 'time']

# Plot the data
plt.figure(figsize=(12, 8))
plt.plot(data_norm['step'], np.log(data_norm['time-step']), label='Level 2 Refinement, Normalized', color='green')
plt.plot(data_3ref['step'], np.log(data_3ref['time-step']), label='Level 2 Refinement', color='red')
plt.plot(data_2ref['step'], np.log(data_2ref['time-step']), label='Level 1 Refinement', color='blue')
plt.xlabel('Number of Steps', fontsize=14)
plt.ylabel('log(Time Step [s])', fontsize=14)
plt.title('Time Step Evolution during Simulation', fontsize=16)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=12, loc="center right")
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.plot(data_norm['time'], (data_norm['time-step']), label='Level 2 Refinement, Normalized', color='green')
plt.plot(data_3ref['time'], (data_3ref['time-step']), label='Level 2 Refinement', color='red')
plt.plot(data_2ref['time'], (data_2ref['time-step']), label='Level 1 Refinement', color='blue')
plt.xlabel('Time [s]', fontsize=14)
plt.ylabel('Time Step [s]', fontsize=14)
plt.title('Time Step Evolution during Simulation', fontsize=16)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=12)
plt.ylim(-0.01, 0.1)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```
