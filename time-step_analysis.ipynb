## Introduction

In this notebook we will plot the time-step evolutions during the performed simulations to see whether they reflect the system's physical dynamics or not. Specifically, we will compare the time-step evolutions during simulations performed with different grid resolutions and different time-stepping strategies (with and without the normalization factor).

```python
# Import the necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Import the csv file with the data from the simulation
data_200_ref3 = pd.read_csv('data/Re200_norm_level2/time_steps.csv')
data_100_ref3 = pd.read_csv('data/Re100_norm_level2/time_steps.csv')
data_50_ref3 = pd.read_csv('data/Re50_norm_level2/time_steps.csv')
data_100_ref2 = pd.read_csv('data/Re100_norm_level1/time_steps.csv')

# Remove the first row of the data frames
data_200_ref3 = data_200_ref3.iloc[1:]
data_100_ref3 = data_100_ref3.iloc[1:]
data_50_ref3 = data_50_ref3.iloc[1:]
data_100_ref2 = data_100_ref2.iloc[1:]

# Add labels to the columns
data_200_ref3.columns = ['step', 'time-step', 'time']
data_100_ref3.columns = ['step', 'time-step', 'time']
data_50_ref3.columns = ['step', 'time-step', 'time']
data_100_ref2.columns = ['step', 'time-step', 'time']

# Plot the data
plt.figure(figsize=(12, 8))
plt.plot(data_200_ref3['step'], np.log(data_200_ref3['time-step']), label='Re=200, Level 2 Refinement', color='green')
plt.plot(data_100_ref3['step'], np.log(data_100_ref3['time-step']), label='Re=100, Level 2 Refinement', color='red')
plt.plot(data_50_ref3['step'], np.log(data_50_ref3['time-step']), label='Re=50, Level 2 Refinement', color='orange')
plt.plot(data_100_ref2['step'], np.log(data_100_ref2['time-step']), label='Re=100, Level 1 Refinement', color='blue')
plt.xlabel('Number of Steps', fontsize=14)
plt.ylabel('log(Time Step [s])', fontsize=14)
plt.title('Time Step Evolution during Simulation', fontsize=16)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=12, loc="center right")
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.plot(data_200_ref3['time'], (data_200_ref3['time-step']), label='Re=200, Level 2 Refinement', color='green')
plt.plot(data_100_ref3['time'], (data_100_ref3['time-step']), label='Re=100, Level 2 Refinement', color='red')
plt.plot(data_50_ref3['time'], (data_50_ref3['time-step']), label='Re=50, Level 2 Refinement', color='orange')
plt.plot(data_100_ref2['time'], (data_100_ref2['time-step']), label='Re=100, Level 1 Refinement', color='blue')
plt.xlabel('Time [s]', fontsize=14)
plt.ylabel('Time Step [s]', fontsize=14)
plt.title('Time Step Evolution during Simulation', fontsize=16)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=12)
plt.ylim(-0.01, 0.12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```
